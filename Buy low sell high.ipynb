```python
def remove_outliers(arr, percentile):
    arr = arr[arr > np.percentile(arr, percentile)]
    arr = arr[arr < np.percentile(arr, 100 - percentile)]
    return arr
    
    
def df_to_return_array(df, percentile_cutoff=1):
    '''
    convert return dataframe to np array
    remove nan and inf values
    remove outliers based on percentile
    '''
    r = df.values.reshape(-1)
    r = r[~np.isnan(r)]
    r = r[~np.isinf(r)]
    r = remove_outliers(r, percentile_cutoff)
    return r
```

```python
def get_dates_below_target(df_price, target):
    '''
    Return dates where daily return is less than target 
    '''
    df_date = df_price.apply(lambda x: x.index)
    df_r = df_price / df_price.shift(1) - 1
    dates_below_target = df_date[df_r < target]
    dates_below_target.reset_index(drop=True, inplace=True)
    dates_below_target = dates_below_target.apply(lambda x: pd.Series(x.dropna(axis=0).values), axis=0)
    return dates_below_target
```

```python
%%time
dates_below_target = get_dates_below_target(df, -0.4)
dates_below_target = dates_below_target.dropna(axis=1, thresh=1)
```

```python
def calculate_r(symbol, investment_date, r_df, holding_period):
    '''
    Calculate return based on holding period (days)
    '''
    if pd.isnull(investment_date):
        return None
    
#     remove empty dates
#     market closed based on local holiday, which is different for some countries
    r_df = r_df.loc[:, [symbol]]
    r_df.dropna(inplace=True)
    
    index = r_df.reset_index()[r_df.reset_index().date == investment_date].index[0]
    daily_r = r_df.iloc[index+1:index+holding_period+1, r_df.columns==symbol]
#     return daily_r
#     return daily_r
    period_r = np.product(1 + daily_r.values) - 1
    return period_r

def calc_r_for_series(s, r_df, holding_period):
    '''
    Calculate return for Series of dates
    '''
    symbol = s.name
    result = s.apply(lambda x: calculate_r(symbol, x, r_df, holding_period))
    return result
```
